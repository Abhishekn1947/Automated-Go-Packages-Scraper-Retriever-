```python
import openpyxl
import requests

# Define your personal access token
access_token = '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'

# Load the Excel file
excel_file = 'C:/Users/abhia/OneDrive/Desktop/Securin intern notes/output.xlsx'
workbook = openpyxl.load_workbook(excel_file)
sheet = workbook.active

# Define the required fields
required_fields = ['name', 'description', 'default_branch']

# Iterate through each row and retrieve package information
for row_index, row in enumerate(sheet.iter_rows(min_row=2, values_only=True), start=2):
    github_link = row[0]

    # Add schema to the URL if it's missing
    if not github_link.startswith(('http://', 'https://')):
        github_link = 'http://' + github_link

    github_api_url = github_link.replace("github.com", "api.github.com/repos")

    try:
        headers = {
            'Authorization': f'Bearer {access_token}',
        }
        response = requests.get(github_api_url, headers=headers)
        response.raise_for_status()

        try:
            repo_info = response.json()

            # Extract the desired information from the repo_info dictionary
            package_info = {}
            for field in required_fields:
                package_info[field] = repo_info.get(field, 'NULL')

            license_info = repo_info.get('license', {})
            package_info['license'] = license_info.get('name', 'NULL') if license_info else 'NULL'

            dependencies_count = repo_info.get('dependencies_count', 0)
            dependents_count = repo_info.get('dependents_count', 0)
            dependent_repositories_count = repo_info.get('dependent_repositories_count', 0)
            total_releases = repo_info.get('total_releases', 0)
            latest_release = repo_info.get('latest_release', {}).get('published_at', 'NULL')
            first_release = repo_info.get('first_release', {}).get('published_at', 'NULL')
            stars = repo_info.get('stargazers_count', 0)
            forks = repo_info.get('forks_count', 0)
            watchers = repo_info.get('subscribers_count', 0)
            contributors = repo_info.get('contributors_count', 0)

            # Print the package information
            print("Package Information:")
            for field, value in package_info.items():
                print(f"{field.capitalize()}: {value}")
            print("Dependencies:", dependencies_count)
            print("Dependent Packages:", dependents_count)
            print("Dependent Repositories:", dependent_repositories_count)
            print("Total Releases:", total_releases)
            print("Latest Release:", latest_release)
            print("First Release:", first_release)
            print("Stars:", stars)
            print("Forks:", forks)
            print("Watchers:", watchers)
            print("Contributors:", contributors)
            print("-----------------------------------------")

            # Update the Excel sheet with the package information
            for col_index, field in enumerate(required_fields, start=2):
                sheet.cell(row=row_index, column=col_index).value = package_info[field]
            sheet.cell(row=row_index, column=5).value = package_info['license']
            sheet.cell(row=row_index, column=6).value = dependencies_count
            sheet.cell(row=row_index, column=7).value = dependents_count
            sheet.cell(row=row_index, column=8).value = dependent_repositories_count
            sheet.cell(row=row_index, column=9).value = total_releases
            sheet.cell(row=row_index, column=10).value = latest_release
            sheet.cell(row=row_index, column=11).value = first_release
            sheet.cell(row=row_index, column=12).value = stars
            sheet.cell(row=row_index, column=13).value = forks
            sheet.cell(row=row_index, column=14).value = watchers
            sheet.cell(row=row_index, column=15).value = contributors

        except ValueError as e:
            print(f"Error parsing JSON response: {e}")
            # Set NULL values in the Excel sheet if the data is not available
            for col_index, field in enumerate(required_fields, start=2):
                sheet.cell(row=row_index, column=col_index).value = 'NULL'
            sheet.cell(row=row_index, column=5).value = 'NULL'
            sheet.cell(row=row_index, column=6).value = 'NULL'
            sheet.cell(row=row_index, column=7).value = 'NULL'
            sheet.cell(row=row_index, column=8).value = 'NULL'
            sheet.cell(row=row_index, column=9).value = 'NULL'
            sheet.cell(row=row_index, column=10).value = 'NULL'
            sheet.cell(row=row_index, column=11).value = 'NULL'

    except requests.exceptions.HTTPError as e:
        print(f"Failed to retrieve package information for {github_link}: {e}")
        # Set NULL values in the Excel sheet if the data is not available
        for col_index, field in enumerate(required_fields, start=2):
            sheet.cell(row=row_index, column=col_index).value = 'NULL'
        sheet.cell(row=row_index, column=5).value = 'NULL'
        sheet.cell(row=row_index, column=6).value = 'NULL'
        sheet.cell(row=row_index, column=7).value = 'NULL'
        sheet.cell(row=row_index, column=8).value = 'NULL'
        sheet.cell(row=row_index, column=9).value = 'NULL'
        sheet.cell(row=row_index, column=10).value = 'NULL'
        sheet.cell(row=row_index, column=11).value = 'NULL'

# Save the modified Excel file
modified_file = 'output.xlsx'
workbook.save(modified_file)
print(f"Package information saved to {modified_file}")
```
